### 抓取猫眼电影的图片和评论

#### 第一步：从已经抓取的html页面中抓取每个电影的url链接

```python
%%time
import urllib
import re
import threading

# 编写单个线程处理函数
def spider_url(filename):
    with open('猫眼电影_1','r') as f:
        test_html = f.read()

    left_url = 'https://maoyan.com'
    pattern = re.compile(r'.*?<p class="name"><a href="(.*?)" title', re.S)
    right_url_list = pattern.findall(test_html)

    for right_url in right_url_list:
        url_list.append(left_url + right_url)

url_list = []
thread_list = []
# 读取文件并分析其中的页面
for i in range(1,11):
    filename = '猫眼电影_{}'.format(i)
    t = threading.Thread(target=spider_url,args=(filename))
    thread_list.append(t)
    t.start()

# 释放线程资源
for t in thread_list:
    t.join()

print(url_list)
```
