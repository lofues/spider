```python
import requests

class ProxyPool(object):
    def __init__(self):
        self.url = ''
        self.headers = {'User-Agent':'Mozilla/5.0'}
        
    def save_proxy(self):
        html = requests.get(url=self.url,headers=self.headers).text
        print(html)
        ip_list = html.split('\r\n')
        print(ip_list)
        with open('/home/tarena/proxies.txt','a') as f:       
            for ip in ip_list:
                # 测试ip,可用的保存到文件
                if self.test_ip(ip):
                    # 把ip写入到文件
                    f.write(ip+'\n')
    
    def test_ip(self,ip):
        proxies = {'http':'http://{}'.format(),'https':'https://{}'.format()}
        try:
            res = requests.get(url='http://www.baidu.com',proxies=proxies,timeout=2)
            if res.status_code == 200:
                
                return True
        except Exception as e:
            print(ip,'不可用')
            return False
            
        
spider = ProxyPool()
spider.save_proxy()
```

```python
import requests

# url = 'http://xicidaili.com/nn/'
url = 'http://www.baidu.com'
headers = {'User-Agent':'Mozilla/5.0'}

proxies_list = []
with open('/home/tarena/secret_ip.txt','r') as f:
    for line in f:
        proxies_list.append(f.readline().strip())
# print(proxies_list)

def test(proxy):
    try:
        res = requests.get(url=url,
                        proxies=proxy,
                        headers=headers,
                        timeout=3
               )
        if res.status_code == 200:
            return True
        else:
            return False
    except Exception as e:
        print(e)

for proxy in proxies_list:
    proxies = {
        'http':'http://{}',
        'https':'http://{}'
    }
    proxies['http'].format(proxy)
    proxies['https'].format(proxy)
    with open('/home/tarena/useable.txt','w') as f:
        if test(proxies):
            print(proxy,'可用')
            f.write(proxy + '\n')
        else:
            print(proxy,'不可用')
```
