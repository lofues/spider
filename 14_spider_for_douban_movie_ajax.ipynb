```python
%%time
import requests
import json
from fake_useragent import UserAgent
import time
import random
from lxml import etree

class DoubanSpider(object):
    def __init__(self):
        self.url = 'https://movie.douban.com/j/chart/top_list?'
        self.list = []
        self.dict = {}
        
    def parse_html(self,params):
        html = requests.get(
            url = self.url,
            params=params,
            headers={'User-Agent':UserAgent().random}
        ).text
        # json.loads(): 将json数据->Python数据类型
        html = json.loads(html)
        for film in html:
            score = film['score']
            name = film['title']
            self.list.append((name,score))
                
    def get_total(self,movie_type):
        params_count = {
                'type': movie_type,
                'interval_id': '100:90',
            }
        json_obj = requests.get(
            url = 'https://movie.douban.com/j/chart/top_list_count?',
            params=params_count,
            headers={'User-Agent':UserAgent().random}
        ).json()
#         print(json_obj)
        count = int(json_obj['total'])
        print('电影总数为:{}'.format(count))
        return count+1
    
    def get_dict_type(self):
        url = 'https://movie.douban.com/chart'
        html = requests.get(url=url,headers={'User-Agent':UserAgent().random}).text
        xpath_bds = "//div[@class='types']/span/a/@href"
        xpath_html = etree.HTML(html)
        url_list = xpath_html.xpath(xpath_bds)
        for url in url_list:
            name = url.split('&')[0].split('?')[1][10:]
            name_type = url.split('&')[1][5:]
            self.dict[name] = name_type
        
            
            
    def run(self):
        self.get_dict_type()
        movie_class = input('请输入你要看的电影类型:')
        movie_type = self.dict.get(movie_class,'')
        if not movie_type:
            print('没有你要看的电影类型')
            return
        total = self.get_total(movie_type)
        print('{}排行榜内的所有电影如下:'.format(movie_class))
        for page in range(0,total,20):
            start = page
            params = {
                'type': movie_type,
                'interval_id': '100:90',
                'action': '',
                'start': str(start),
                'limit': '20'
            }
            self.parse_html(params)
            time.sleep(random.random())
        print(self.list)
        
spider = DoubanSpider()
spider.run()    
```
